# Análise de Chip de Memória DRAM (4M x 1)

Este notebook resolve a questão sobre a organização interna e multiplexação de endereços de um chip de 4 Mbits.

```python
import math

# Dados do chip
capacidade_bits = 4 * 1024 * 1024  # 4096K = 4M
pinos_endereco = 11

# Cálculo de endereçamento sem multiplexação (linear)
enderecos_necessarios = math.log2(capacidade_bits)

# Cálculo com multiplexação (Matriz Linha x Coluna)
# 11 bits para Linha (RAS) e 11 bits para Coluna (CAS)
linhas = 2**pinos_endereco
colunas = 2**pinos_endereco
total_posicoes = linhas * colunas

print(f"Capacidade Total: {capacidade_bits} bits")
print(f"Pinos de endereço físicos: {pinos_endereco}")
print(f"Endereços possíveis via multiplexação (2^{pinos_endereco} * 2^{pinos_endereco}): {total_posicoes}")

if total_posicoes == capacidade_bits:
    print("\nConclusão: A multiplexação permite acessar todas as posições usando apenas 11 pinos.")
```

### Respostas Teóricas

**a) Como um barramento de 11 pinos consegue acessar 4.096K posições?**
Isso ocorre através da técnica de **multiplexação de endereços**. O endereço total de 22 bits (necessário para $2^{22} = 4.194.304$ posições) é dividido em duas partes de 11 bits. Primeiro, os 11 pinos recebem o endereço da **linha** (row) e, em um segundo momento, os mesmos 11 pinos recebem o endereço da **coluna** (column). Isso reduz o custo físico do chip ao diminuir a quantidade de pinos necessários.

**b) Qual é a função das linhas de controle $\overline{\text{RAS}}$ e $\overline{\text{CAS}}$?**
*   **$\overline{\text{RAS}}$ (Row Address Strobe):** Sinaliza ao chip que os bits presentes nos pinos $A_0$ a $A_{10}$ representam o endereço da **linha** na matriz de memória. Ele trava (latch) esse endereço internamente.
*   **$\overline{\text{CAS}}$ (Column Address Strobe):** Sinaliza que os bits nos pinos representam o endereço da **coluna**. A intersecção da linha selecionada pelo RAS com a coluna selecionada pelo CAS identifica a célula de 1 bit a ser acessada.
